### <font color = "#757168">Effect of Socio Economic variables on Consumer price index and Consumer confidence index</font>

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='job',y='cons.price.idx',data=newdf)
plt.title('Boxplot of consumer price index with respect to job',fontsize=20)
plt.xticks(rotation= 45)
plt.show()
```

The boxplot indicate that the average value of price indices differs among the clients with different jobs

Clients having their jobs as Housemaids and not mentioned have more consitent consumer price indices and much variation is observed in the case of students and retired clients.

The least median value(average) is for the students

Next we do One way Analysis of Variance to find out the average value for the consumer price index differs between different job or not


```python
stats.f_oneway(df['job'],df['cons.price.idx'] ,axis=0)
```

Since the calculated p value is less than 0.05 we conclude that the average value of the consumer price index significantly differs between different jobs

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='job',y='cons.conf.idx',data=newdf)
plt.title('Boxplot of consumer condifidence index with respect to job',fontsize=20)
plt.xticks(rotation= 45)
plt.show()
```

The boxplot indicate that the average statisfaction indices differs among clients with different job

Clients having their jobs in services, blue collar and students has wider variations than the other jobs in the case consumer confidence index

Retired has the highest median value

The ANOVA carried out to test whether there is any significant difference exist for consumer confidence index among different job category


```python
stats.f_oneway(df['job'],df['cons.conf.idx'] ,axis=0)
```

Since the calculated p value is less than 0.05 we conclude that significant difference exist for consumer confidence index among different job category

```python
plt.figure(figsize=(15,7))
cmatrix = newdf.corr()
sns.heatmap(cmatrix,annot=True,linewidths=0.9,linecolor="#ffa929",cmap="flare")
plt.show()
```

The correlation indicate that consumer confidence index is independent of all other qualitative variables but in the case of consumer price index are correlated with emp.var.rate,euribor3m,nr.employed so we find the carry out the Linear regression to find the contribution of these variables on consumer price index


```python
X1= newdf[['emp.var.rate','euribor3m','nr.employed']]
```

```python
y1=pd.DataFrame(newdf['cons.price.idx'])
```

```python
from sklearn.model_selection import train_test_split
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,random_state=42,test_size=.25)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(X1_train,y1_train)
pred=model.predict(X1_test)
```

```python
from sklearn.metrics import mean_squared_error
print("MSE is",mean_squared_error(y1_test,pred))
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
from sklearn.metrics import r2_score
r2_score(pred,y1_test)
```

Regression equation is

cons.price.idx= 135.47443043 + 0.63416746 * emp.var.rate + -0.00910256 * euribor3m + -0.00811239 * nr.employed

From the equation it is easy to infer that emp variation rate is the most contributing factor to consumer price index


# <font color = "#ff8000">Modelling</font>

```python
df.columns
```

```python
y = df['y']
x = df.drop(['y'],axis = 1)
print("Shape of X is : ",x.shape)
print("Shape of Y is : ",y.shape)
```

```python
x.head()
```

```python
y.head()
```

```python
# Splitting the Data
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(xtrain,ytrain)

ypred = lr.predict(xtest)
```

```python
from sklearn.metrics import f1_score,accuracy_score

print("F1 Score is : ",f1_score(ypred,ytest, average = 'micro'))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [    
    {'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : np.logspace(-4, 4, 20),
    'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter' : [1000,2500, 5000,10000]
    }
]
```

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
acc_values = []
neighbors = np.arange(3,15)
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k, metric='minkowski')
    classifier.fit(xtrain, ytrain)
    ypred =  classifier.predict(xtest)
    acc = accuracy_score(ytest, ypred)
    acc_values.append(acc)
```

```python
plt.plot(neighbors,acc_values,'o-')
plt.xlabel("K Value")
plt.ylabel("Accuracy")
```

```python
classifier = KNeighborsClassifier(n_neighbors = 12, metric='minkowski')
classifier.fit(xtrain, ytrain)
ypred =  classifier.predict(xtest) 
```

```python
print("F1 Score is : ",f1_score(ypred,ytest, average = 'micro'))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
models = [LogisticRegression(max_iter=1000),KNeighborsClassifier(),DecisionTreeClassifier(),
         RandomForestClassifier(), SVC()]
f1_score = []

for model in models:
    cv = cross_val_score(model,xtrain,ytrain,scoring="f1_micro",cv = 10).mean()
    f1_score.append(cv)
    
```

```python
model = ["Logistic Regression", "KNN", "Decision Tree","Random Forest","SVC"]

res = pd.DataFrame({'Model':model,'F1 Score':f1_score})
res
```

Logistic Regression has the highest F1 score

# <font color = "#439c49">Fine Tuning</font>

```python
lr = LogisticRegression()
param_grid = {
    'C':np.logspace(-5, 8, 15),
    'max_iter':[1000]
}
lr_search = GridSearchCV(lr, param_grid=param_grid, refit = True, verbose = 3, cv=5)


lr_search.fit(xtrain , ytrain)
lr_search.best_params_
```

```python
print("Tuned Logistic Regression Parameters: {}".format(lr_search.best_params_)) 
print("Best score is {}".format(lr_search.best_score_))
```
